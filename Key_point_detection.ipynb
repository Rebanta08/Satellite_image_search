```
import numpy as np
import cv2
import os
```

```
surf = cv2.xfeatures2d.SIFT_create(400)

indexList=[]
files=[]

des_all = None
for index, filename in enumerate(os.listdir('/content/drive/MyDrive/Image_Search/database')):
    img2 = cv2.imread('/content/drive/MyDrive/Image_Search/database/' + filename, 0)
    kp2, des2 = surf.detectAndCompute(img2, None)
    if des_all is None:
        des_all = des2
    else:
        des_all = np.concatenate((des_all, des2))
    indexList.extend([index]*des2.shape[0])
    files.append(filename)
    print(filename)
```

```
scanned='/content/drive/MyDrive/Image_Search/queries/query1.png'
img = cv2.imread(scanned, 0)
kp1, des1 = surf.detectAndCompute(img, None)

scanned='/content/drive/MyDrive/Image_Search/queries/query2.png'
img = cv2.imread(scanned, 0)
kp2, des2 = surf.detectAndCompute(img, None)

scanned='/content/drive/MyDrive/Image_Search/queries/query3.png'
img = cv2.imread(scanned, 0)
kp3, des3 = surf.detectAndCompute(img, None)

scanned='/content/drive/MyDrive/Image_Search/queries/query4.png'
img = cv2.imread(scanned, 0)
kp4, des4 = surf.detectAndCompute(img, None)

scanned='/content/drive/MyDrive/Image_Search/queries/query5.png'
img = cv2.imread(scanned, 0)
kp5, des5 = surf.detectAndCompute(img, None)

print(des_all.shape)
print(des3.shape)
```

```
FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
flann = cv2.flann_Index()

print ("Training...")
flann.build(des_all, index_params)

print ("Matching 1st image...")
indexes1, matches1 = flann.knnSearch(des1, 10)
print ("Matching 2nd image...")
indexes2, matches2 = flann.knnSearch(des2, 10)
print ("Matching 3rd image...")
indexes3, matches3 = flann.knnSearch(des3, 10)
print ("Matching 4th image...")
indexes4, matches4 = flann.knnSearch(des4, 10)
print ("Matching 5th image...")
indexes5, matches5 = flann.knnSearch(des5, 10)
```

```
[m1,n1]=indexes1.shape
a1=np.reshape(indexes1,(m1*n1))

[m2,n2]=indexes2.shape
a2=np.reshape(indexes2,(m2*n2))

[m3,n3]=indexes3.shape
a3=np.reshape(indexes3,(m3*n3))

[m4,n4]=indexes4.shape
a4=np.reshape(indexes4,(m4*n4))

[m5,n5]=indexes5.shape
a5=np.reshape(indexes5,(m5*n5))
```

```
from collections import Counter
from scipy import stats
h=np.array(indexList)

votes1 = h[a1]
#stats.mode(votes1)
likelyMatch1 = np.bincount(votes1).argmax()

votes2 = h[a2]
#stats.mode(votes2)
likelyMatch2 = np.bincount(votes2).argmax()

votes3 = h[a3]
#likelyMatch3 = stats.mode(votes3)
likelyMatch3 = np.bincount(votes3).argmax()

votes4 = h[a4]
#likelyMatch4 = stats.mode(votes4)
likelyMatch4 = np.bincount(votes4).argmax()

votes5 = h[a5]
#likelyMatch5 = stats.mode(votes5)
likelyMatch5 = np.bincount(votes5).argmax()
```

```
image_num_1 = files[likelyMatch1].split("e")
image_1=image_num_1[1].split('.')
likely1='image'+str(int(image_1[0])+525)+'.png'
print(likely1)

image_num_2 = files[likelyMatch2].split("e")
image_2=image_num_2[1].split('.')
likely2='image'+str(int(image_2[0])+525)+'.png'
print(likely2)

image_num_3 = files[likelyMatch3].split("e")
image_3=image_num_3[1].split('.')
likely3='image'+str(int(image_3[0])+525)+'.png'
print(likely3)

image_num_4 = files[likelyMatch4].split("e")
image_4=image_num_4[1].split('.')
likely4='image'+str(int(image_4[0])+525)+'.png'
print(likely4)

image_num_5 = files[likelyMatch5].split("e")
image_5=image_num_5[1].split('.')
likely5='image'+str(int(image_5[0])+525)+'.png'
print(likely5)
```
